```python
import sqlite3
import json
import os
```

```python
DB_PATH = "/home/jkr/work/open-science/open-science/test_db/index.db"
LIMIT = 100
print(f"Does file exist? {os.path.exists(DB_PATH)}")

extracted_links = []
with sqlite3.connect(DB_PATH) as conn:
  cursor = conn.cursor()
  cursor.execute(f"SELECT oa_urls FROM works WHERE oa_urls IS NOT NULL AND oa_urls != '' LIMIT {LIMIT}")
  results = cursor.fetchall()
  for row in results:
    json_str = row[0]
    oa_data = json.loads(json_str)
    extracted_links.extend(oa_data['pdf_links'])
  conn.close()

print(extracted_links[:10])
```
